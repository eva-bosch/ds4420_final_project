```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import math
import csv
```

<br></br>

```python
df = pd.read_csv('user_tea_pairs.csv', delimiter=',', header=None)
df
```

<br></br>

<br></br>

```python
all_tea_data = df.values.tolist()
```

# Get Tea Data

```python
num_teas = len(all_tea_data)
```

```python
for ind in range(num_teas):
    try:
        # get data from url
        url = all_tea_data[ind][1]
        r = requests.get(url)
        soup = BeautifulSoup(r.content, 'html.parser')
        
        # get ratings
        tables = soup.find_all('table')
        tab2 = tables[2]
        rate = list(map(lambda x: x.text.split()[0], tab2.find_all('td')[4:8]))
        
        # get tea info
        tab1 = tables[1]
        info = tab1.find_all('td')
        char = list(map(lambda x: x.text, info[1:11:2]))
        
        # get product link
        lnk = info[-1].a['href']
        
        all_tea_data[ind] += rate
        all_tea_data[ind] += char
        all_tea_data[ind].append(lnk)

        # check if added correct number of characteristics
        if (len(rate) + len(char)) != 9:
            print(f'Tea {ind} added {len(rate) + len(char) + 1} instead of {9} new parameters')
        
        if ind % 500 == 0:
            print(f'Tea {ind} added successfully; {ind} / {num_teas}')
    except:
        print(f'Encounter issue with tea {ind}')
```

<br></br>

# Save

```python
column_names = ['Account Id', 'Review Link', 'Tea Name', 'Aroma Rating', 'Flavor Rating', 
                'Value Rating', 'Total Rating', 'Brand', 'Style', 'Region', 'Caffeine', 'Loose Type', 'Product page']
```

```python
all_tea_data.insert(0, column_names)
```

```python
with open('tea_data.csv', 'w', encoding='utf-8') as f:
    write = csv.writer(f)
    write.writerows(all_tea_data)
```
